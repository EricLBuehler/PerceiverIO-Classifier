```python
import matplotlib.pyplot as plt
from google.colab import drive
import os
```

```python
modelname="2_4_22_m1"

drive.mount('/content/drive')

prefix='/content/drive/MyDrive/Colab Notebooks/PercieverIO_Classifier/'

try:
    os.mkdir(prefix)
except FileExistsError:
    pass

prefix_models=prefix+'models/'+modelname+"/"
try:
    os.mkdir(prefix+'models/')
except FileExistsError:
    pass
try:
    os.mkdir(prefix_models)
except FileExistsError:
    pass
```

```python
if len(os.listdir(prefix_models))>0:
    dir=[item for item in os.listdir(prefix_models) if ".txt" in item]
    

    with open(prefix_models+"train_loss.txt","r") as file:
        fileread=file.read().split("\n")
        fileread.pop()
        #Hack to quickly get losses out
        train_loss=[float(item.split("_")[1]) for item in fileread]

    with open(prefix_models+"test_loss.txt","r") as file:
        fileread=file.read().split("\n")
        fileread.pop()
        #Hack to quickly get losses out
        test_loss=[float(item.split("_")[1]) for item in fileread]

    with open(prefix_models+"accuracy.txt","r") as file:
        fileread=file.read().split("\n")
        fileread.pop()
        #Hack to quickly get losses out
        accuracy=[float(item.split("_")[1]) for item in fileread]

    plt.plot(range(len(train_loss)),train_loss,color="red")
    plt.title("Train loss")
    plt.xlabel("Batch")
    plt.ylabel("Loss")
    plt.savefig(prefix_models+"train_loss.png")
    plt.show()
    
    plt.plot(range(len(test_loss)),test_loss,color="orange")
    plt.title("Test loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.savefig(prefix_models+"test_loss.png")
    plt.show()

    
    plt.plot(range(len(accuracy)),accuracy)
    plt.title("Accuracy")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.savefig(prefix_models+"accuracy.png")
    plt.show()


```
